```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('textract_dataframe.csv')
df.sort_values('score',ascending=False).tail(20)
```

```python
hand_data = df.groupby('hand').score.mean()

plt.style.use('ggplot')
plt.bar(hand_data.argsort(),hand_data,color='darkblue',width=0.5)
plt.xticks(hand_data.argsort(),hand_data.keys())

for i,v in zip(hand_data.argsort(),hand_data):
    plt.text(i-0.2,v-0.1,f'$\mu={round(v,3)}$',fontdict={'color':'white','size':10})

plt.show()
```

```python
gender_data = df.groupby('gender').score.mean()

# plt.style.use('ggplot')
plt.bar(gender_data.argsort(),gender_data,color=['pink','lightblue'],width=0.5)
plt.xticks(gender_data.argsort(),gender_data.keys())

for i,v in zip(gender_data.argsort(),gender_data):
    plt.text(i-0.175,v-0.1,f'$\mu={round(v,3)}$',fontdict={'color':'white','size':14})

plt.show()
```
