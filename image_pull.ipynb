```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import urllib
import boto3
import string
from io import StringIO, BytesIO
import json
import os
from PIL import Image
import requests
from IPython.display import display
```

```python
def display_sample(sample_id):
    url = f'https://project-350-warner.s3.amazonaws.com/AllHandwritingImages/{sample_id}.png'
    resp = requests.get(url)
    img = Image.open(BytesIO(resp.content))
    img = img.resize((img.width//5,img.height//5))
    display(img)
```

```python
df = pd.read_csv('textract_dataframe.csv')
```

```python
num =  np.random.randint(0,df.index.max())
display_sample(df.id[num])
```
